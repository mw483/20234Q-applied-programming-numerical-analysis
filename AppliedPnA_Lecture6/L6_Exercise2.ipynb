```python
def student_rank(lab_choice, lab_name, student_no):
    for i, student in enumerate(lab_choice[lab_name]):
        if student_no == student:
            return i
    return -1


def galeShapley(lab_capacity, lab_choice, student_choice):
    n_student = len(student_choice)
    n_lab = len(lab_choice)

    lab_assignment = {}
    for i in lab_choice.keys():
        lab_assignment[i] = []

    selected = {}
    for i in student_choice.keys():
        selected[i] = 0

    count = 1
    for i in range(n_lab):
        allStudentsSelected = True
        for s in selected:
            if s != 0:
                allStudentsSelected = False
        if allStudentsSelected:
            break
    
        for _ in student_choice:
            for j in student_choice.keys():
                if selected[j] or i >= len(student_choice[j]):
                    continue
    
                preferred_lab = student_choice[j][i]
                studentRank = student_rank(lab_choice, preferred_lab, j)
    
                if studentRank == -1:
                    continue
    
                if len(lab_assignment[preferred_lab]) < lab_capacity[preferred_lab]:
                    lab_assignment[preferred_lab].append(studentRank)
                    selected[j] = preferred_lab
                else:
                    max_rank_in_lab = max(lab_assignment[preferred_lab])
                    if studentRank < max_rank_in_lab:
                        lab_assignment[preferred_lab].remove(max_rank_in_lab)
                        selected[lab_choice[preferred_lab][max_rank_in_lab]] = 0
                        lab_assignment[preferred_lab].append(studentRank)
                        selected[j] = preferred_lab
        print(f'Tentative lab assignment round {count}')
        print('-----------------------------------------')
        print(f'Unassigned students:', end=' ')
        for k,v in selected.items():
            if v == 0:
                print(k, end=', ')
        print()
        print('Assigned students: ')
        for k in lab_assignment.keys():
            print(k, end=': ')
            for lab_student_rank in lab_assignment[k]:
                print(lab_choice[k][lab_student_rank], end=', ')
            print()
        print()
        count +=1


lab_capacity = {'l1': 1, 'l2': 1, 'l3': 1}
lab_choice = {'l1': ['s5', 's3', 's2', 's6', 's4'],
              'l2': ['s2', 's6', 's3', 's5'],
              'l3': ['s2', 's1', 's3', 's6', 's4', 's5']}
student_choice = {'s1': ['l1'],
                  's2': ['l1', 'l3'],
                  's3': ['l2', 'l1', 'l3'],
                  's4': ['l1', 'l2', 'l3'],
                  's5': ['l3', 'l2', 'l1'],
                  's6': ['l1', 'l2', 'l3']}

galeShapley(lab_capacity, lab_choice, student_choice)
```
